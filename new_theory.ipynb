```python
def high_low(x, low, high, threshold=0.3):

    if x < -1 and high < threshold:
        return True
    elif x > 1 and low > -threshold:
        return True
    elif x < -1 and high > threshold:
        return False
    elif x > 1 and low < -threshold:
        return False
    else:
        return 'neutral'
    
def calculate_profit(per, low, high, high_first, low_first):

    if high_first == 1:
        if per < -0.3:
            return -0.6
        else:
            return per - 0.3
    elif low_first == 1:
        if per > 0.3:
            return -0.6
        else:
            return -1*(per + 0.3)
    else:
        return 0
```

```python
df['Month'] = df['time_buy'].dt.strftime('%m')
df['day'] = df['time_buy'].dt.strftime('%d')
```

```python
df['result'] = df.apply(lambda row: high_low(row['%'], row['low'], row['high']), axis=1)
df['profit'] = df.apply(lambda row: calculate_profit(row['%'], row['low'], row['high'], row['high_first'],row['low_first']), axis=1)
df['profit$'] = (df.profit / 100) * 10000
```
